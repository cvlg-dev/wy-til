```r
library(stats)
```

```r
crime = read.csv('data/USArrests.csv')
rownames(crime) <- crime$States
tail(crime)
```

```r
data = crime[, 2:5]
summary(data)
```

```r
data_pca = princomp(data, cor=T, scores=T)
data_pca
```

```r
summary(data_pca)
```

```r
screeplot(data_pca, type='lines', pch=19, main='Scree Plot')
```

```r
data_pca_scores = data_pca$scores[, 1:2]
data_pca_scores
```
