```r
options(repr.plot.width = 15, repr.plot.height = 10)
```

```r
data <- ts(read.csv("data/gdpq.csv", header=T), start=1970, frequency=4)
gdp = data[, 1]
gdpsa = data[, 2]
```

```r
plot(gdp/1000, ylab="GDP", xlab="YEAR", col="black")
lines(gdpsa/1000,  col="red")
```

```r
# 5  
ww = c(.5, 1, 1, 1, .5)
gdpm5 = filter(gdp, sides=2, ww/sum(ww))
```

```r
dlgdp1 = diff(log(gdp))  #  1
dlgdp4 = diff(log(gdp), 4)  #  4
```

```r
par(mfrow=c(2,2))

plot(gdp, main="GDP", ylab=" ", xlab="YEAR")
plot(gdpm5, main="GDPM5", ylab=" ", xlab="YEAR")
plot(dlgdp1, main="diff(log(GDP))", ylab=" ", xlab="YEAR")
plot(dlgdp4, main="diff(log(GDP), 4)", ylab=" ", xlab="YEAR")
```
