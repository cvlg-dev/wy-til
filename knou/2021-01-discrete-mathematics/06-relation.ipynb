# 6.1. 관계 기본사항

## 6.1.1. 정의

### 6.1.1.1. 곱집합 (Cartesian Product)

- 집합 A와 B의 곱집합은 $A \times B$
- A의 원소와 B의 원소의 모든 순서쌍(ordered pair)들의 집합
    - $A \times B = \{(a, b) | a \in A, b\in B$
    
### 6.1.1.2. 관계 (Relation)

- 집합 X에서 집합 Y로의 **(이항)관계( binary relation)** R은  $X \times Y$의 부분집합
- $(x, y) \in R$
    - x는 y와 R의 관계가 있다.
    - $xRy$로 표기
    - X=Y 일 경우에는 "R을 X에서의 관계"라고 표현함

# 6.2. 관계의 표현


## 6.2.1. 화살표 도표

- X에서 Y로의 관계 R
    - $(x, y) \in R$
- 화살표도표는 위의 조건을 만족하는 원소끼리만 화살표로 연결


## 6.2.2. 방향 그래프

- 점(vertex), 선(edge)로 이루어진 도형
- $G = (V, E)$
    - 엣지가 방향을 가지면 G는 방향 그래프
    - 방향을 가지지 않으면 G는 무향 그래프
    
    
## 6.2.3. 부울 행렬

- $m \times n$의 부울행렬 $M_R = (a_{ij})$
    - 행렬을 구성하고 관계가 있으면 1, 없으면 0 으로 표현

# 6.3. 관계의 성질

## 6.3.1. 성질의 종류

1. 반사적 reflective
    - 어떠한 원소여도 동일한 값이 존재해야 한다.
    - $\forall a \in A, (a, a) \in R$
2. 대칭적 symmetric
    - $\forall a, b \in A, (a, b) \in R \Rightarrow (b, a) \in R$
3. 추이적 transitive
    - $\forall a, b, c, \in A, ((a, b) \in R \wedge (b, c) \in R) \Rightarrow (a, c) \in R$

## 6.3.2. 관계의 성질과 부울 행렬

### 1. 반사적

- $\forall a \in A, (a, a) \in R$
- $M_R$은 대각행렬

### 2. 대칭적

- $\forall a, b \in A, (a, b) \in R \Rightarrow (b, a) \in R$
- $M_R = M^T_R$ 즉, 대칭행렬


### 3. 추이적

- $\forall a, b, c \in A, ((a, b) \in R \wedge (b, c) \in R) \Rightarrow (a, c) \in R$

# 6.4. 관계의 종류



## 6.4.1. 역관계 (inverse relation)
```
X, Y: 집합
R: X에서 Y로의 관계
```
- $R^{-1}$: R의 역관계
    - $R^{-1} = \{(y, x) | (x, y) \in R\} \subset Y \times X$


## 6.4.2. 합성관계 (composition relation)

```
A, B, C: 집합
R: X에서 Y로의 관계
S: B에서 C로의 관계
```

- R과 S의 합성관계 
- $S \circ R = \{(a, c) | a \in A, b \in B, c \in C, (a, b) \in R, (b, c) \in S \}$
- $S \circ R \subset A \times C$

### 6.4.2.1. 합성관계의 부울행렬 표현

```
A, B, C: 유한집합 (|A| = m, |B| = n, |C| = p 
R: X에서 Y로의 관계
S: B에서 C로의 관계
```

- $S \circ R$: A에서 C로의 관계
    - $M_R$은 $m \times n$ 의 부울행렬
    - $M_S$은 $n \times p$ 의 부울행렬
    - $M_{S \circ R}$은 $m \times p$ 의 부울행렬
    - $M_{S \circ R} = M_R \odot M_S$


## 6.4.3. 동치관계 (equivalent relation)

```
R: 집합 A에서 관계
```

- R이 반사적, 대칭적, 추이적의 성질을 다 만족한다면,
    - R은 동치관계

## 6.4.4. 동치류 (equivalence class)

```
A: 집합
R: A에서의 동치관계
```

- A의 임의의 원소 a에 대해서
    - A의 원소들 x중, a로 시작해서 x로 끝나는 원소들을 a의 동치류라고 함.
    - $[a] = \{x \in A\ | (a, x) \in R\}$
