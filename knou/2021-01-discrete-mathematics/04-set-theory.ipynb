# 4.1. 집합 기본 사항

## 4.1.1. 논리학과 집합론

논리명제에서 이야기 하는 것과 집합론은 유사하게 볼 수 있음
- 논리합 $p(x) \vee g(x) \rightarrow A \cup B$ 합집합
- 논리곱 $p(x) \wedge g(x) \rightarrow A \cap B$ 교집합

## 4.1.2. 집합과 원소

집합은 기본적으로 무정의용어 

- 무정의용어: 정의없이 사용하는 용어. 
    - 직관적으로 이해할 수 있으나 다른 용어로 정의하기 힘든 대상을 표현하기 위해 사용됨

## 4.1.3. 집합의 표기법

- a는 S의 원소
    - $a \in S$

- S는 중괄호 { }로 표기
    - 원소나열법
    - 조건나열법
    
- 집합의 크기
    - $|S| = 3$

## 4.1.4. 부분집합

- 정의 
    - A의 모든 원소가 B의 원소이면 A는 B의 부분집합
    - $A \subseteq B$ 또는 $A \subset B$
    - $\forall x ... |(x \in A \rightarrow x \in B)$

### 4.1.4.1. 진부분집합 (proper subset)

- A는 B의 진부분집합
    - $A \subseteq B, B \subseteq / A$
    - $A \subseteq B, A \neq B$
    
### 4.1.4.2. 상동 (equal)

- A = B라는 것은
    - $A \subseteq B, B \subseteq A$

## 4.1.5. 서로소 (disjoint)

- 두 집합이 교집합이 공집합일 때, 서로소
    - 집합A와 B는 서로소
    - $A \cap B = \emptyset$
- 쌍으로 서로소 (pairwise disjoint)
    - n개의 집합 $A_n$은, 집합들 간의 원소가 그 어느 것도 교집합이 없음
    - $A_i \cup A_j = \emptyset (i \neq j)$

## 4.1.6. 분할 (partition)

- 집합 A을 공집합이 아닌 부분집합들로 나눌 때, 
- A의 모든 원소들이 각각 나누어진 부분집합들 중 하나에만 포함될 경우
- 부분집ㅎ바들 전체의 집합을 A의 분할이라고 함.
    - partition of A = ${A_1, A_2, A_3 ... A_n}$
    
1. $ \cup^n_{i=1} A_i = A $
2. $ \cap^n_{i=1} A_i = \emptyset$

## 4.1.7. 멱집합

- 집합 A의 모든 부분집합들의 집합을 A의 멱집합
    - $P(A)$
    - 공집합도 반드시 포함됨
    
- 멱집합 원소의 수
    - $|S| = n$
    - $|P(S)| = 2^n$
