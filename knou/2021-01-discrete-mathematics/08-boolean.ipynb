# 8.1. 부울대수 소개

## 8.1.1. 디지털 논리회로

디지털신호가 입력되었을 때, 논리연산을 거쳐 디지털신호를 출력하는 과정

- 현대 디지털 컴퓨터의 수학적 구조는 이진법
    - 이진법: 0과 1로만 모든 수를 표현하는 수 체계

## 8.1.2. 기본 논리 게이트

### 8.1.2.1. AND 게이트

- $F = X \cdot Y$ 
    - 논리연산 논리곱 : $p \wedge q$

### 8.1.2.2. OR 게이트

- $F = X + Y$ 
    - 논리연산 논리합 : $p \vee q$

### 8.1.2.3. NOT 게이트

- $F = \bar{X}$
    - 논리연산 : $\sim p$

## 8.1.3. 기타 논리 게이트


### 8.1.3.1. NAND 게이트

`AND & NOT`

- $F = \bar{X \cdot Y} = \bar{X} + \bar{Y}$


### 8.1.3.2. NOR 게이트

`OR & NOT`

- $F = \bar{X + Y} = \bar{X} \cdot \bar{Y}$


### 8.1.3.3. XOR 게이트

`Exclusive OR (배타적 논리합)` 

- $F = X \oplus Y = \bar{X}Y + X\bar{Y}$


### 8.1.3.3. XNOR 게이트

`Exclusive OR + NOT` 

- $F = \bar{X \oplus Y} = \bar{\bar{X}Y + X\bar{Y}} = \bar{X}\bar{Y} + XY$

# 8.2. 부울대수

## 8.2.1. 부울식

- 부울식의 순환적 정의
    - 부울상수 0, 1은 부울식
    - 부울변수는 부울식
    - X, Y가 부울식일 때, X, Y를 활용한 논리연산도 부울식
    
## 8.2.2. 부울대수의 성질


### 8.2.2.1. 성질 

1. $ X + 0 = X $ : OR
2. $ X \cdot 1 = X $ : AND
3. $ X + 1 = 1$
4. $X \cdot 0 = 0$
5. $X + X = X$
6. $X \cdot X = 0$
7. $X + \bar{X} = 1$
8. $X \cdot \bar{X} = 0$
9. $\bar{\bar{X}}=X$
10. $X+Y = Y+X$
11. $XY = YX$
12. $X + (Y+Z) = (X+Y)+Z$
13. $X(YZ) = (XY)Z$
14. $X(Y+Z) = XY + XZ$
15. $X+YZ = (X+Y)(X+Z)$
16. $\bar{X+Y} = \bar{X} \cdot \bar{Y}$
17. $\bar{X \cdot Y} = \bar{X} + \bar{Y}$
18. $X+X \cdot Y = X$
19. $X \cdot (X + Y) = X$

### 8.2.2.2. 논리연산과 부울대수의 관련성

- 기호
    - $p, q, r \Longleftrightarrow X, Y, Z$
    - $T \Longleftrightarrow 1$
    - $F \Longleftrightarrow 0$
    - $AND \wedge \Longleftrightarrow \cdot$
    - $OR \vee \Longleftrightarrow + $
    - $NOT \sim \Longleftrightarrow \bar{}$
    
    
- 법칙
    - 항등법칙
        - $p \vee F \equiv p \Longrightarrow X + 0 = X$
        - $p \wedge T \equiv p \Longrightarrow X \cdot 1 = X$
    - 지배법칙
        - $p \vee T \equiv T \Longrightarrow X + 1 = 1$
        - $p \wedge F \equiv F \Longrightarrow X \cdot 0 = 0$
    - 멱등법칙
        - $p \vee p \equiv p \Longrightarrow X + X = X$
        - $p \wedge p \equiv p \Longrightarrow X \cdot X = X$    
    - 부정법칙
        - $p \vee (\sim p) \equiv T \Longrightarrow X + \bar{X} = 1$
        - $p \wedge (\sim p) \equiv F \Longrightarrow X \cdot \bar{X} = 0$    
        - $\sim (\sim p) \equiv p \Longrightarrow \bar{\bar{X}} = X$
    - 교환법칙
        - $p \vee q \sim q \vee p \Longrightarrow X + Y = Y + X$
        - $p \wedge q \sim q \wedge p \Longrightarrow X \cdot Y = Y \cdot X$
    - 결합법칙
        - $p \vee (q \vee r) \equiv (p \vee q) \vee r \Longrightarrow X + (Y+Z) = (X+Y) + Z$
        - $p \wedge (q \wedge r) \equiv (p \wedge q) \wedge r \Longrightarrow X \cdot (Y \cdot Z) = (X \cdot Y) \cdot Z$
    - 분배법칙
        - $p \wedge (q \vee r) \equiv (p \wedge q) \vee (p \wedge r) \Longrightarrow X \cdot (Y+Z) = X \cdot Y + X \cdot Z$
        - $p \vee (q \wedge r) \equiv (p \vee q) \wedge (p \vee r) \Longrightarrow X + (Y \cdot Z) = (X + Y) \cdot (X + Z)$
    - 드모르간 법칙
        - $\sim (p \vee q) \equiv (\sim p) \wedge (\sim q) \Longrightarrow \bar{X + Y} = \bar{X} \cdot \bar{Y}$
        - $\sim (p \wedge q) \equiv (\sim p) \vee (\sim q) \Longrightarrow \bar{X \cdot Y} = \bar{X} + \bar{Y}$        
    - 흡수법칙
        - $p \vee (p \wedge q) \equiv p \Longrightarrow X+X \cdot Y = X$
        - $p \wedge (p \vee q) \equiv p \Longrightarrow X \cdot (X + Y) = X$

## 8.2.3. 쌍대성 원리 및 보수

### 8.2.3.1. 쌍대성 원리 (principle of duality)

- 쌍대
    - 부울식에서 논리곱과 논리합을 서로 바꾸고, 
    - 논리상수 0과 1을 서로 바꾸면
        - 원래 부울식의 쌍대(dual) 을 얻게 됨
    - 주어진 부울식과 그것의 쌍대는 진리값이 서로 같음

    1. $ X + 0 = X \Longrightarrow X \cdot 1 = X $
    2. $ X + 1 = 1 \Longrightarrow X \cdot 0 = 0$
    3. $X + X = X \Longrightarrow X \cdot X = 0$
    4. $X + \bar{X} = 1 \Longrightarrow X \cdot \bar{X} = 0$
    5. $X+Y = Y+X \Longrightarrow XY = YX$
    6. $X + (Y+Z) = (X+Y)+Z \Longrightarrow X(YZ) = (XY)Z$
    7. $X(Y+Z) = XY + XZ \Longrightarrow X+YZ = (X+Y)(X+Z)$
    8. $\bar{X+Y} = \bar{X} \cdot \bar{Y} \Longrightarrow \bar{X \cdot Y} = \bar{X} + \bar{Y}$
    9. $X+X \cdot Y = X \Longrightarrow X \cdot (X + Y) = X$
    

- 드모르간 법칙의 일반화 

- 부울 함수의 보수
    - 부울함수 $F$의 보수는 $\bar{F}$
- 보수 구하기
    - 드모르간 법칙
    - 쌍대성 원리 이용
        - $F$의 쌍대를 구함
        - $F$의 쌍대에서 정상형($X$)와 보수형 ($\bar{X}$)을 서로 바꿈

# 8.3. 부울함수의 대수적 간소화


## 8.3.1. 간소화의 목적

- 복잡한 논리회로란 복잡한 부울함수로 인한 것
    - 따라서 간단한 부울함수를 통해 간소화된 논리회로와 동일한 기능을 꾀하는 것이 바람직함
    
예)
- $ F = $
    - $\bar{X}YZ + X \bar{Y} Z + XYZ $
    - $\bar{X}YZ + X \bar{Y} Z + XYZ + XYZ$
    - $YZ(\bar{X} + X) + XZ(\bar{Y}+Y)$
    - $YZ \cdot 1 + XZ \cdot 1$
    - $YZ + XZ$

## 8.3.2. 간소화의 예시

1. 항결합
    - 두 개의 항을 결합하여 하나의 항으로 만드는 방법
    - 결합법칙 사용
2. 문자 소거
    - 중복된 문자를 제거하는 방법
    - 흡수법칙 사용
3. 중복함 첨가
    - 진리값이 변하지 않도록 하면서 간소화를 위한 적절한 항을 첨가하는 방법
    - 예)
        - $F = F + X\bar{X}$
        - $F = F \cdot (X + \bar{X}$
        - $F + X = F + X + XY$ ... (흡수법칙)
        - $F + X = F + X \cdot (X+Y)$
